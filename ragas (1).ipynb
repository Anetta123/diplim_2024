```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
# Путь к файлу CSV
csv_file_path = 'data.csv'
# Чтение данных из CSV-файла
df = pd.read_csv(csv_file_path)
# Выбор столбцов для heatmap
heatmap_data = df[['answer_relevancy', 'answer_relevancy', 'context_recall', 'faithfulness', 'context_recall']]
# Создание кастомного colormap
cmap = LinearSegmentedColormap.from_list('green_red', ['red', 'green'])
# Настройка размера графика
plt.figure(figsize=(10, 8))
# Построение heatmap
sns.heatmap(heatmap_data, annot=True, fmt=".2f", linewidths=.5, cmap=cmap)

# Установка меток по оси y (если необходимо)
# Предполагается, что в вашем CSV-файле есть столбец 'question' с метками
plt.yticks(ticks=range(len(df['question'])), labels=df['question'], rotation=0)
# Отображение графика
plt.show()
```
